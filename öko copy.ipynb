```python
import pandas as pd
df = pd.read_csv('train.csv')

df.head()
```

```python
df.describe().transpose()
```

```python
df_corr = df.corr()['SalePrice'].sort_values().to_frame()
df_corr
```

```python
# Selecting columns with lower than 0.5 correlation
columns_to_drop = df_corr[df_corr['SalePrice'] < 0.5].transpose().columns
```

```python
# Dropping the columns_to_drop from the dataframe
df_cleaned = df.drop(columns_to_drop, axis=1)
```

```python
# Selecting the categorical columns
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
categorical_columns
```

```python
five_types = ['ExterQual', 'ExterCond', 'HeatingQC', 'KitchenQual', 'PoolQC']
six_types = ['BsmtQual', 'FireplaceQu', 'GarageQual', 'GarageCond']
```

```python
# Scoring from 0 to 5 the columns with 5 types
for column in five_types:
    df_cleaned[column] = df_cleaned[column].map({
        'Ex': 5,
        'Gd': 4,
        'TA': 3,
        'Fa': 2,
        'Po': 1,
        'NA': 0
    })
```

```python
# Scoring from 0 to 6 the columns with 6 types
for column in six_types:
    df_cleaned[column] = df_cleaned[column].map({
        'Ex': 6,
        'Gd': 5,
        'TA': 4,
        'Fa': 3,
        'Po': 2,
        'NA': 1
    })
```

```python
df_cleaned[six_types].head()
```

```python
df_cleaned.columns
```

```python
frequency_encoding = ['MSZoning', 'LandContour', 'Neighborhood', 'Exterior1st', 'Exterior2nd']
# Frequency encoding
for column in frequency_encoding:
    # Calculate frequency of each category in the column
    frequency_map = df_cleaned[column].value_counts(normalize=True)
    
    # Replace the original values with their corresponding frequencies
    df_cleaned[column] = df_cleaned[column].map(frequency_map)
```

```python
df.describe().transpose()
```

```python
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
```

```python
# Clear NaN values
df_cleaned = df_cleaned.fillna(0)

# Dropping categorical_columns
df_cleaned = df_cleaned.drop(categorical_columns, axis=1)
```

```python
# Multiple Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df_cleaned.drop('SalePrice', axis=1)
y = df_cleaned['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = LinearRegression()
model.fit(X_train, y_train)

# Predicting the test set
y_pred = model.predict(X_test)
```

```python
# Calculating the R2 score
from sklearn.metrics import r2_score

# Calculating the RMSE
from sklearn.metrics import mean_squared_error

print('R2 score: ', r2_score(y_test, y_pred))
print('RMSE: ', mean_squared_error(y_test, y_pred, squared=False))
```

```python
# Printing the model equation
print('y = ', model.intercept_, end='')
for i in range(len(model.coef_)):
    print('\n + ', model.coef_[i], ' * ', X.columns[i], end='')
print()
```

```python
# Plotting the residuals
import matplotlib.pyplot as plt
import seaborn as sns
residuals = y_test - y_pred

sns.scatterplot(x=y_test, y=residuals)
plt.axhline(y=0, color='red', ls='--')

```
